```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import joblib 
```

```python
#importation de la dataset
dataset=pd.read_excel('DATASET_MOTO.xls')
dataset
```

```python
#determiner les x et les y
x=dataset[['marque','acheteur','venteur','qualite_moto','prix',]]
y=dataset['prix']
```

```python
#on subdivise la dataset en 2 parties
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size = 0.2, random_state=0)
x_train
```

```python
#choisir l'algorithme
regression=LinearRegression()

#entrainer grace au modele fit
regression.fit(x_train,y_train)
```

```python
#enregistrer le model
joblib.dump(regression, 'modelemot.pkl')

#faire le test
y_pred=regression.predict(x_test)
```

```python
df=pd.DataFrame({'prix actuel' :y_test, 'prix predites' :y_pred})
df
```
